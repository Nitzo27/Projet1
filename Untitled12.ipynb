```python

# Bounding box function
def bou_box30(x,y):
  
  h = max(y) - min(y)
  w = max(x) - min(x)
  xminp = min(x)-0.3*w ; xmaxp = max(x)+0.3*w
  yminp = min(y)-0.3*h ; ymaxp = max(y)+0.3*h
  return xminp,xmaxp,yminp,ymaxp
# Bounding box function 30%
def bou_box(x,y):
  xmax=max(x)
  xmin=min(x)
  ymax=max(y)
  ymin=min(y)
  return xmin,xmax,ymin,ymax
```
